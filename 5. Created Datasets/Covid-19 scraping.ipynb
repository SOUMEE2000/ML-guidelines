```python
url="https://www.mohfw.gov.in/"
```

```python

from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get(url)
soup= BeautifulSoup(page.content,'html.parser')
```

```python
soup.prettify()
```

```python
data=soup.select('table.table-striped ')
data
```

```python
countries=[]
confirmed=[]
cured=[]
death=[]

for i in range(1, len(data), 5):
    countries.append(data[i].get_text())
for i in range(2, len(data), 5):
    confirmed.append(data[i].get_text())
for i in range(3, len(data), 5):
    cured.append(data[i].get_text())
for i in range(4, len(data), 5):
    death.append(data[i].get_text())

    
```

```python
countries
```

```python
countries= countries[0:-3]
death=death[:-2]
confirmed= confirmed[:-3]
cured= cured[:-3]
```

```python
import pandas as pd
df= pd.DataFrame()
df["Countries"]=countries
df["Death"]=death
df["Confirmed"]=confirmed
df["Cured"]=cured
df.head()
```
